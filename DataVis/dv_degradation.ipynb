```python
import numpy as np
import matplotlib.pyplot as plt
from math import pi
df = pd.read_csv(r'C:\Users\gajalakshmi\OneDrive\Documents\projects\monuments\temple.csv')
numeric_df = df.select_dtypes(include=[np.number])
data = numeric_df.iloc[0]
categories = data.index
values = data.values
values = np.array(values, dtype=float)
N = len(categories)
angles = [n / float(N) * 2 * pi for n in range(N)]
values = np.concatenate((values, [values[0]])) 
angles += angles[:1]  
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))
ax.set_theta_offset(pi / 2)
ax.set_theta_direction(-1)
plt.xticks(angles[:-1], categories, color='grey', size=8)
ax.plot(angles, values, color='red', linewidth=2, linestyle='solid')
ax.fill(angles, values, color='red', alpha=0.4)
plt.title('Radar Chart')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
df = pd.read_csv(r'C:\Users\gajalakshmi\OneDrive\Documents\projects\monuments\temple.csv')
numeric_df = df.select_dtypes(include=[np.number])
df_melted = numeric_df.melt()
plt.figure(figsize=(12, 8))
sns.violinplot(x='variable', y='value', data=df_melted)
plt.xticks(rotation=90)
plt.title('Violin Plot')
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
df = pd.read_csv(r'C:\Users\gajalakshmi\OneDrive\Documents\projects\monuments\temple.csv')
x = df['Temperature_Max (°C)']
y = df['Humidity (%)']
z = df['Rainfall (mm)']
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, z, color='b', marker='o')
ax.set_xlabel('Temperature_Max (°C)')
ax.set_ylabel('Humidity (%)')
ax.set_zlabel('Rainfall (mm)')
ax.set_title('3D Scatter Plot')
plt.show()
```
